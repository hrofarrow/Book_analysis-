```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("google_books_1299.csv").dropna()
data = data.drop_duplicates(subset ="title", 
                     keep = "first", inplace = False)
data.head()
```

```python
data['published_date'] = data['published_date'].astype('datetime64[ns]')
data["month"] = data.published_date.dt.month
data
```

```python
data['month'].replace(12,0, inplace = True)
data = pd.DataFrame(data)
data
```

```python
labels = ['Winter', 'Spring', "Summer", 'Fall']
bins = [-.01,2.99,5.99,8.99,11]
data['season'] = pd.cut(data['month'], bins=bins, labels=labels)
data.head(20)
```

```python
season_count = data['season'].value_counts().sort_index().plot(kind='bar',color = ['orange', 'green' , 'red', 'blue'], ylim = (30,75))
season_count.set_ylabel("No. of books written")
season_count.set_xlabel("Season")
season_count.set_title("Books written per Season")
```

```python
season_price = data.groupby('season')['price'].mean().sort_index().plot(kind='bar', ylim = (30,50), color = ['orange', 'green' , 'red', 'blue'])
season_price.set_ylabel("Average Price of a Book")
season_price.set_xlabel("Season")
season_price.set_title("Average Price of a Book per Season")
```
